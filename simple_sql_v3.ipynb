```python
%%sql
--creating view
CREATE VIEW new_view AS
SELECT 'Video ID', channelTitle, title, viewCount, likeCount, commentCount, duration
FROM channels;
```

```python
%%sql
--checking if veiw is made
SELECT *
FROM sqlite_schema;
```

```python
%%sql 
SELECT COUNT(*)
FROM new_view;
```

```python
%%sql 
SELECT DISTINCT channelTitle
FROM new_view
LIMIT 20;
```

```python
%%sql 
SELECT channelTitle, count(*) AS 'Numbers of Videos'
FROM new_view
WHERE channelTitle LIKE 'ABC NEWS%' or channelTitle LIKE 'BBC%' or channelTitle like 'CNN%'
GROUP by channelTitle;
```

```python
%%sql 
SELECT channelTitle, count(*) AS 'Numbers of Videos'
FROM new_view
GROUP by channelTitle;
```

```python
%%sql 
SELECT title as 'Video Name', channelTitle as 'Channel Name', MAX(viewCount) as 'Highest Numbers of Views'
FROM new_view
LIMIT 5;
```

```python
%%sql 
SELECT title as 'Video Name', channelTitle as 'Channel Name', viewCount as 'Lowest Numbers of Views'
FROM new_view
ORDER BY viewCount ASC
LIMIT 5;
```

```python
%%sql 
SELECT channelTitle, SUM(viewCount) AS 'Total Amount of views'
FROM new_view
WHERE channelTitle LIKE 'USA Today%'
GROUP by channelTitle;
```

```python
%%sql 
SELECT channelTitle, AVG(likeCount) AS 'Amount Amount of like Count'
FROM new_view
WHERE channelTitle LIKE 'CBS News%';
```

```python
%%sql 
SELECT channelTitle, SUM(viewCount) AS 'Total Amount of views', SUM(likeCount) AS 'Total Amount of Likes', SUM(CommentCount) AS 'Total Amount of Comments'
FROM new_view
GROUP BY channelTitle
ORDER BY SUM(viewCount) DESC;
```

```python
%%sql 
SELECT channelTitle, SUM(viewCount) AS 'Total Amount of views', SUM(likeCount) AS 'Total Amount of Likes', SUM(CommentCount) AS 'Total Amount of Comments'
FROM new_view
GROUP BY channelTitle
ORDER BY SUM(viewCount) DESC;
```

```python
%%sql 
SELECT channelTitle, SUM(viewCount) AS 'Total Amount of views'
FROM new_view
GROUP BY channelTitle
HAVING SUM(viewCount) > 100000000
ORDER BY SUM(viewCount) ASC;
```

```python
%%sql

SELECT ROUND((SUM(likeCount) *100)/(SUM(viewCount)),2) AS 'Overall Percentage of People who likes the Videos after Watching'
FROM new_view;
```
