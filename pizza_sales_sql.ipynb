```python
%%sql
SELECT quantity, price
FROM orders
LEFT JOIN orderDetails
    ON orders.order_id = orderDetails.order_id
LEFT JOIN pizzas
    ON orderDetails.pizza_id = pizzas.pizza_id
Where quantity > 1
LIMIT 5;
```

```python
%%sql


--CREATE VIEW q1a AS
-- top 20 dates with the highest income
SELECT date as 'Date', SUM(quantity) as 'Amount of Pizza Sold', ROUND(SUM(price*quantity),2) as 'Total'
FROM orderDetails
LEFT JOIN orders
    ON orders.order_id = orderDetails.order_id
LEFT JOIN pizzas
    ON orderDetails.pizza_id = pizzas.pizza_id
GROUP BY date
ORDER BY Total DESC
LIMIT 5;
```

```python
%%sql
SELECT * 
FROM q1a;
```

```python
%%sql
CREATE VIEW q1b AS
-- top 20 dates with the LOWEST income
SELECT date as 'Date', SUM(quantity) as 'Amount of Pizza Sold', ROUND(SUM(price*quantity),2) as 'Total'
FROM orderDetails
LEFT JOIN orders
    ON orders.order_id = orderDetails.order_id
LEFT JOIN pizzas
    ON orderDetails.pizza_id = pizzas.pizza_id
GROUP BY date
ORDER BY Total ASC
LIMIT 20;
```

```python
%%sql

CREATE VIEW q2 AS
--Hours of the day that are most busy and less busy
SELECT strftime('%H', time) as 'Hours', SUM(quantity) as 'Amount of Pizza Sold', ROUND(SUM(price*quantity),2) as 'Total'
FROM orderDetails
LEFT JOIN orders
    ON orders.order_id = orderDetails.order_id
LEFT JOIN pizzas
    ON orderDetails.pizza_id = pizzas.pizza_id
GROUP by Hours
LIMIT 50;
```

```python
%%sql

CREATE VIEW q3 AS
--Most Popular pizza and their size ordering by Total
SELECT pizzaTypes.name, pizzas.size, ROUND(SUM(price*quantity),2) as 'Total', SUM(quantity) as 'Amount of Pizza Sold' 
FROM orderDetails
LEFT JOIN order
    ON orders.order_id = orderDetails.order_id
LEFT JOIN pizzas
    ON orderDetails.pizza_id = pizzas.pizza_id
LEFT JOIN pizzaTypes
    ON pizzas.pizza_type_id = pizzaTypes.pizza_type_id
GROUP BY orderDetails.pizza_id
ORDER BY Total DESC;
```

```python
%%sql

CREATE VIEW q4 AS
SELECT  
CASE 
    WHEN strftime('%m', date) == '01' or strftime('%m', date)== '02' or strftime('%m', date) == '12' THEN 'Winter'
    WHEN strftime('%m', date) == '03' or strftime('%m', date) == '04' or strftime('%m', date) == '05' THEN 'Spring'
    WHEN strftime('%m', date) == '06' or strftime('%m', date) == '07' or strftime('%m', date) == '08' THEN 'Autumn'
    WHEN strftime('%m', date) == '09' or strftime('%m', date) == '10' or strftime('%m', date) == '11' THEN 'Fall'
END AS Season, SUM(quantity) as 'Amount of Pizza Sold', ROUND(SUM(price*quantity),2) as 'Total'
FROM orderDetails
LEFT JOIN orders
    ON orders.order_id = orderDetails.order_id
LEFT JOIN pizzas
    ON orderDetails.pizza_id = pizzas.pizza_id
GROUP BY Season
ORDER BY Total DESC;
```
