```python
%%sql

SELECT sex,AVG(age)
FROM people
GROUP By sex;
```

```python
%%sql

SELECT Owner_name as "Richest Home Owner", Area, MAX(value) as 'Highest Priced Home'
FROM homes;
```

```python
%%sql

SELECT Area, MAX(value) as 'Highest Value'
FROM homes
GROUP BY Area;
```

```python
%%sql

SELECT Area,AVG(value) as 'Average Home Value'
FROM homes
GROUP BY area;
```

```python
%%sql

SELECT Area,ROUND(AVG(value),2) as 'Average Home Value'
FROM homes
GROUP BY area;
```

```python
%%sql
SELECT *
FROM people
JOIN pets
ON pets.owner_name = people.name;
```

```python
%%sql
SELECT *
FROM people
JOIN pets
ON pets.owner_name = people.name
WHERE type = 'cat';
```

```python
%%sql

SELECT DISTINCT(owner_name) AS OWNER, age
FROM pets
LEFT JOIN people
ON pets.owner_name = people.name;
```

```python
%%sql

SELECT homes.owner_name, area, name
FROM pets
JOIN homes
ON pets.owner_name = homes.owner_name
```

```python
%%sql

SELECT *
FROM pets
JOIN homes
ON pets.owner_name = homes.owner_name
GROUP BY name
HAVING homes.value > 150000
ORDER by area;
```
